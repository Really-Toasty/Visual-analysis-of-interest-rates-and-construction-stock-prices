```python
index = merged_df.copy()
```

```python
index
```

```python
index['BBY.L_index'] = (index['BBY.L'] / 195.681900) *100
index['BKG.L_index'] = (index['BKG.L'] / 202.653763) *100
index['IBST.L_index']= (index['IBST.L'] / 195.423996) *100
index['KIE.L_index']= (index['KIE.L'] / 86.63682556) *100
index['KLR.L_index']= (index['KLR.L'] / 128.6818695) *100
index['MSLH.L_index'] = (index['MSLH.L'] / 70.613358) *100
index['PSN.L_index'] = (index['PSN.L'] / 137.990082) *100
index['TW.L_index'] = (index['TW.L'] / 207.049393) *100

index['interest_index'] = (index['interest'] / 9.72) *100
```

```python
index
```

```python
plt.figure(figsize=(12, 6))
index[['BBY.L_index', 'BKG.L_index', 'IBST.L_index', 'KIE.L_index', 'KLR.L_index', 'MSLH.L_index', 'PSN.L_index', 'TW.L_index', 'interest_index']].plot()
plt.title('test graph')
plt.legend(['BBY.L', 'BKG.L_index', 'IBST.L_index', 'KIE.L_index', 'KLR.L_index', 'MSLH.L_index', 'PSN.L_index', 'TW.L_index', 'Interest Rate'])
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Index Value')

plt.show()
```

```python
stock_index = index[['Date', 'BBY.L_index', 'BKG.L_index', 'IBST.L_index', 'KIE.L_index', 'KLR.L_index', 'MSLH.L_index', 'PSN.L_index', 'TW.L_index']]
stock_index
```

```python
stock_index.columns = ['Date', 'Balfour Beatty', 'Berkeley Group Holdings', 'Ibstock', 'Kier Group', 'Keller Group', 'Marshalls', 'Persimmon', 'Taylor Wimpey']
stock_index
```

```python
# Loop through each stock and plot individually with the interest rate
for stock in stock_index:
    fig, ax = plt.subplots(figsize=(12, 6))  # Create a new figure for each stock
    ax.plot(index['Date'], index[stock], label=stock, color='blue')  # Plot stock prices
    ax.plot(index['Date'], index['interest_index'], label='Interest Rate', color='green')  # Plot interest rate
    ax.set_title(f'{stock} and Interest Rate')
    ax.set_xlabel('Date')
    ax.set_ylabel('Value')
    ax.grid(True)
    ax.legend()
    
    plt.show()  # Display each plot one by one
```
